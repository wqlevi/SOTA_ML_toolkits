```python
from torch import nn
from PIL import Image
import torch
import matplotlib.pyplot as plt
from torchvision.transforms.functional import pil_to_tensor as ptt
```

```python
def image_to_tensor(im)->torch.Tensor:
    tmp_tensor = ptt(im)
    return tmp_tensor[0][None,None,:,:].to(torch.float32)
```

```python
conv_pad_0 = nn.Sequential(nn.Conv2d(1,1,3,2,1),
                          nn.ReLU()
)

conv_pad_circle = nn.Sequential(nn.Conv2d(1,1,3,2,1,padding_mode='circular'),
                               nn.ReLU()
)
```

```python
def _init_weight(m):
    if isinstance(m, nn.Conv2d):
        m.bias.data.fill_(0)
        m.weight.data.fill_(1)
```

```python
im = Image.open("example.png")
x = image_to_tensor(im)
```

```python
x.shape
```

```python
def pad_img_ts(x)->torch.Tensor:
    x_ = x.clone()
    x_ = torch.cat((x_, x_[:,:,:,-2:-1]),-1)
    return x_
```

```python
x_pad = pad_img_ts(x)
```

```python
conv_pad_0.apply(_init_weight)
conv_pad_circle.apply(_init_weight)
```

```python
y_pad_0, y_pad_cir = conv_pad_0(x), conv_pad_circle(x)
```

### parameters in the conv kernel

$$[255, 255, 255] * [1,1,1] = [2295, 2295, 2295]$$

### You see the 0-padding artifacts in subfig. 1?

```python
min_v = y_pad_0.min().item()
max_v = y_pad_0.max().item()
```

```python
#color scaled to [0,2295], which is min and max of the first image
fig,ax = plt.subplots(1,2,figsize=(50,50))
ax[0].imshow(y_pad_0.squeeze().detach().numpy()[:50,:50],vmin=min_v,vmax=max_v)
ax[1].imshow(y_pad_cir.squeeze().detach().numpy()[:50,:50],vmin=min_v,vmax=max_v)
plt.show()
```

```python
#color scaled to [0,2295], which is min and max of the first image
fig,ax = plt.subplots(1,2,figsize=(50,50))
ax[0].imshow(y_pad_0.squeeze().detach().numpy()[-50:,-50:],vmin=min_v,vmax=max_v)
ax[1].imshow(y_pad_cir.squeeze().detach().numpy()[-50:,-50:],vmin=min_v,vmax=max_v)
plt.show()
```

## pad input to odd dimension
* even dimension induces spatially biased activation(along each row)
* uneven dimension has symmetrical activation(along column) 

```python
x_pad.shape
```

```python
y_pad_0, y_pad_cir = conv_pad_0(x_pad), conv_pad_circle(x_pad)
```

```python
#color scaled to [0,2295], which is min and max of the first image
fig,ax = plt.subplots(1,2,figsize=(50,50))
ax[0].imshow(y_pad_0.squeeze().detach().numpy()[:50,:50],vmin=min_v,vmax=max_v)
ax[1].imshow(y_pad_cir.squeeze().detach().numpy()[:50,:50],vmin=min_v,vmax=max_v)
plt.show()
```
